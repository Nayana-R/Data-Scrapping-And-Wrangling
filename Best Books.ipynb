```python
import pandas as pd
import numpy as np

from selenium import webdriver
browser = webdriver.Chrome('Chromedriver.exe')

url = 'https://www.goodreads.com/choiceawards/best-books-2017'
browser.get(url)
```

```python
books_category = pd.DataFrame()

parent_books = browser.find_element_by_css_selector('div.categoryContainer')
books = parent_books.find_elements_by_css_selector('div.category.clearFix')

for book in books:
    category = book.find_element_by_css_selector('h4.category__copy').text
    name = book.find_element_by_css_selector('img.category__winnerImage').get_attribute('alt')
    image = book.find_element_by_css_selector('img.category__winnerImage').get_attribute('src')
    description = book.find_element_by_css_selector('a').get_attribute('href')

    book_dict = {'Category':category, 'Name':name, 'Image_url':image, 'Description_url':description}
    books_category = books_category.append(book_dict, ignore_index = True)
    
books_category.to_csv('Book Category 1')
books_category
```

```python
import pandas as pd
import numpy as np

books_category1 = pd.DataFrame()

for book1 in range(len(books_category)):
    url_1 = books_category.loc[book1, 'Description_url']
    browser.get(url_1)
    #print(url_1)
    parent1 = browser.find_element_by_css_selector('div.pollContents')
    books1 = parent1.find_element_by_css_selector('div.winningBook')
    
    author = books1.find_element_by_css_selector('span[itemprop="name"]').text
    short_desc = books1.find_element_by_css_selector('div#description.readable').text
    votes = books1.find_element_by_css_selector('span.greyText.gcaNumVotes').text
    total_votes = parent1.find_element_by_css_selector('div.greyText.gcaNumVotes').text
    vote = int((votes.split(" ")[0]).replace(',',''))
    total = int((total_votes.split(" ")[0]).replace(',',''))
    winner_percent_votes = round(((vote/total)*100),2)
    
    book_dict1={'Author':author, 'Description':short_desc, 'Winner votes':votes, 'Total votes': total_votes,
                'Vote Percentage':winner_percent_votes, 'URL':url_1}
    
    books_category1 = books_category1.append(book_dict1, ignore_index=True)

books_category1.to_csv('Book Category 2')
books_category1  
```

```python
book_category_final = pd.merge(books_category, books_category1, left_on='Description_url',right_on='URL')
ls = ['Category', 'Description_url', 'URL', 'Name', 'Author', 'Description', 'Total votes', 'Vote Percentage', 'Winner votes']
book_category_final = book_category_final[ls]
book_category_final.to_csv('Book Category')
book_category_final
```
